```
import torchvision.datasets as dset
import torchvision.transforms as transforms
#resizeしてToTenserしてnormする
trans = transforms.Compose([transforms.Resize((224,224)), transforms.ToTensor(), transforms.Normalize((0.5,), (1.0,))])
train_set = dset.CIFAR10(root="./", train=True, transform=trans, download=True)
test_set = dset.CIFAR10(root="./", train=False, transform=trans, download=True)
```

```
import torch
batch_size = 100

train_loader = torch.utils.data.DataLoader(
                 dataset=train_set,
                 batch_size=batch_size,
                 shuffle=True)
test_loader = torch.utils.data.DataLoader(
                dataset=test_set,
                batch_size=batch_size,
                shuffle=False)
```
