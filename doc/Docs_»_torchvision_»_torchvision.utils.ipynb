# 画像のグリッドを作る

torchvision.utils.make_grid(tensor, nrow=8, padding=2, normalize=False, range=None, scale_each=False, pad_value=0)

````
import torch
import torchvision.transforms as transforms
from torchvision.utils import make_grid, save_image
import matplotlib.pyplot as plt
import numpy as np
import scipy.misc
%matplotlib inline
````

````
def show(img):
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1,2,0)), interpolation='nearest')
````

````
lena = scipy.misc.face()
img = transforms.ToTensor()(lena)
print(img.size())
````

````
show(img)
````

## １行３列のグリッド画像

````
imglist = [img, img, img]
show(make_grid(imglist))
````

## 各行に表示される画像数を指定

nrowに１を設定したした場合は１行の中に１画像を表示。

````
imglist = [img, img, img]
show(make_grid(imglist, nrow=1))
````

nrowに２を設定したした場合は１行の中に２画像を表示。

````
imglist = [img, img, img]
show(make_grid(imglist, nrow=2))
````

## パディング

デフォルトは2。

````
show(make_grid(imglist, padding=100))
````

````
show(make_grid(imglist, padding=200))
````

````
show(make_grid(imglist, padding=300))
````

##正規化

normalize=Trueで画像を0〜1の値に正規化する。

````
show(make_grid(imglist, normalize=True))
````

##  正規化の範囲を指定

````
show(make_grid(imglist, padding=100, normalize=True, range=(0, 1)))
````

````
show(make_grid(imglist, padding=100, normalize=True, range=(0, 0.5)))
````

## バッチ内の画像で正規化

````
show(make_grid(imglist, padding=100, normalize=True, scale_each=True))
````

````
show(make_grid(imglist, padding=100, normalize=True, range=(0, 0.5), scale_each=True))
````

## パディングのピクセル値

````
show(make_grid(imglist, padding=100, pad_value=0))
````

````
show(make_grid(imglist, padding=100, pad_value=100))
````
