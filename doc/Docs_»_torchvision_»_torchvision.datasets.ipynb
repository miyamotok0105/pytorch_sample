# FashionMNIST

```
import torchvision.datasets as dset
import torchvision.transforms as transforms

trans = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5,), (1.0,))])
train_set = dset.FashionMNIST(root="./", train=True, transform=trans, download=True)
test_set = dset.FashionMNIST(root="./", train=False, transform=trans, download=True)
```

```
import torch
batch_size = 100

train_loader = torch.utils.data.DataLoader(
                 dataset=train_set,
                 batch_size=batch_size,
                 shuffle=True)
test_loader = torch.utils.data.DataLoader(
                dataset=test_set,
                batch_size=batch_size,
                shuffle=False)
```

# CIFAR10

```
import torch
batch_size = 100

train_loader = torch.utils.data.DataLoader(
                 dataset=train_set,
                 batch_size=batch_size,
                 shuffle=True)
test_loader = torch.utils.data.DataLoader(
                dataset=test_set,
                batch_size=batch_size,
                shuffle=False)
```

```
for batch_idx, (x, target) in enumerate(train_loader):
    print(x.shape)
    print(x.size())
    print(type(x))
    print(torch.is_tensor(x))
    print(torch.is_storage(x))
    print(target.shape)
    print(target.size())
    print(type(target))
    print(torch.is_tensor(target))
    print(torch.is_storage(target))
    break
```

# resize

https://discuss.pytorch.org/t/resizing-any-simple-direct-way/10316/7

```
import torch
import torch.nn.functional as F
from torch.autograd import Variable

def resize2d(img, size):
    return (F.adaptive_avg_pool2d(Variable(img,volatile=True), size)).data

img = torch.Tensor(3,256,256).uniform_()
print(img.size())
img = resize2d(img, (224,224))
print(img.size())
```
